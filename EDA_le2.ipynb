```python
import pandas as pd
import numpy as np
```

```python
df_gdp = pd.read_csv('data/data.csv')
df_gdp.drop(columns=['drop'], inplace=True)

# replace everything that is not a number with NaN
df_gdp["year"].replace(r'[^\d.]+', np.nan, regex=True, inplace=True)
# drop NaN
df_gdp.dropna(inplace=True)
# turn year into int
df_gdp['year'] = df_gdp['year'].astype(int)
df_gdp['lifeExp'] = df_gdp['lifeExp'].astype(float)
```
