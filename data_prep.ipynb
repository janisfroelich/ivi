```python
import pandas as pd
import numpy as np
```

```python
# read csv from folder data
#flight_data_2014 = pd.read_csv("Data/2014.csv")
#flight_data_2015 = pd.read_csv("Data/2015.csv")
#flight_data_2016 = pd.read_csv("Data/2016.csv")
#flight_data_2017 = pd.read_csv("Data/2017.csv")
flight_data_2018 = pd.read_csv("Data/2018.csv")
airport_data = pd.read_csv("Data/Airports/airports.csv")
```

```python
# combine all df together
#flight_data = pd.concat([flight_data_2017, flight_data_2018], ignore_index=True)
```

```python
flight_data = flight_data_2018
```

```python
flight_data.info()
```

```python
# drop columns that are not needed
flight_data = flight_data.drop(['CARRIER_DELAY', 'WEATHER_DELAY', 'NAS_DELAY', 'SECURITY_DELAY', 'LATE_AIRCRAFT_DELAY', 'Unnamed: 27', 'DIVERTED', 'CANCELLED', 'CANCELLATION_CODE'], axis=1)
```

```python
flight_data.head()
```

```python
# drop all NaN in CRS_ARR_TIME, CRS_DEP_TIME, DEP_TIME, ARR_TIME
flight_data = flight_data.dropna(subset=['CRS_ARR_TIME', 'CRS_DEP_TIME', 'DEP_TIME', 'ARR_TIME'])
```

```python
flight_data[flight_data.DEP_TIME.isna()]
```

```python
flight_data.head()
```

```python
flight_data.DEP_TIME = flight_data.DEP_TIME.astype(int)
flight_data.ARR_TIME = flight_data.ARR_TIME.astype(int)
```

```python
flight_data.dtypes
```

```python
flight_data.CRS_DEP_TIME = flight_data.CRS_DEP_TIME/100
flight_data.DEP_TIME = flight_data.DEP_TIME/100
flight_data.CRS_ARR_TIME = flight_data.CRS_ARR_TIME/100
flight_data.ARR_TIME = flight_data.ARR_TIME/100
```

```python
flight_data.head()
```

```python
# turn CRS_DEP_TIME, CRS_ARR_TIME, DEP_Time, ARR_TIME into string
flight_data['CRS_DEP_TIME'] = flight_data['CRS_DEP_TIME'].astype(str)
flight_data['CRS_ARR_TIME'] = flight_data['CRS_ARR_TIME'].astype(str)
flight_data['DEP_TIME'] = flight_data['DEP_TIME'].astype(str)
flight_data['ARR_TIME'] = flight_data['ARR_TIME'].astype(str)
```

```python
# replace . with : for CRS_DEP_TIME, CRS_ARR_TIME, DEP_Time, ARR_TIME
flight_data['CRS_DEP_TIME'] = flight_data['CRS_DEP_TIME'].str.replace('.', ':')
flight_data['CRS_ARR_TIME'] = flight_data['CRS_ARR_TIME'].str.replace('.', ':')
flight_data['DEP_TIME'] = flight_data['DEP_TIME'].str.replace('.', ':')
flight_data['ARR_TIME'] = flight_data['ARR_TIME'].str.replace('.', ':')
```

```python
flight_data
```

```python
flight_data.dtypes
```

```python
# if length after : is 0 add 00
# if length after : is 1 add 0
flight_data['CRS_DEP_TIME'] = flight_data['CRS_DEP_TIME'].apply(lambda x: x if len(x.split(':')[1]) == 2 else x + '0')
flight_data['CRS_ARR_TIME'] = flight_data['CRS_ARR_TIME'].apply(lambda x: x if len(x.split(':')[1]) == 2 else x + '0')
flight_data['DEP_TIME'] = flight_data['DEP_TIME'].apply(lambda x: x if len(x.split(':')[1]) == 2 else x + '0')
flight_data['ARR_TIME'] = flight_data['ARR_TIME'].apply(lambda x: x if len(x.split(':')[1]) == 2 else x + '0')
```

```python
flight_data
```

```python
# if length of flight_data[CRS_DEP_TIME] < 4 add 0 to the front
flight_data['CRS_DEP_TIME'] = flight_data['CRS_DEP_TIME'].apply(lambda x: '0' + x if len(x) < 5 else x)
flight_data['CRS_ARR_TIME'] = flight_data['CRS_ARR_TIME'].apply(lambda x: '0' + x if len(x) < 5 else x)
flight_data['DEP_TIME'] = flight_data['DEP_TIME'].apply(lambda x: '0' + x if len(x) < 5 else x)
flight_data['ARR_TIME'] = flight_data['ARR_TIME'].apply(lambda x: '0' + x if len(x) < 5 else x)
```

```python
flight_data.head()
```

```python
# display all data types
flight_data.dtypes
```

```python
# if CRS_DEP_TIME, CRS_ARR_TIME, DEP_Time, ARR_TIME is 24:00 change to 00:00
flight_data['CRS_DEP_TIME'] = flight_data['CRS_DEP_TIME'].apply(lambda x: '00:00' if x == '24:00' else x)
flight_data['CRS_ARR_TIME'] = flight_data['CRS_ARR_TIME'].apply(lambda x: '00:00' if x == '24:00' else x)
flight_data['DEP_TIME'] = flight_data['DEP_TIME'].apply(lambda x: '00:00' if x == '24:00' else x)
flight_data['ARR_TIME'] = flight_data['ARR_TIME'].apply(lambda x: '00:00' if x == '24:00' else x)
```

```python
# combine FL_DATE and CRS_DEP_TIME to get DEP_DATETIME as datetime
flight_data['DEP_DATETIME'] = pd.to_datetime(flight_data['FL_DATE'] + ' ' + flight_data['CRS_DEP_TIME'].astype(str), format='%Y-%m-%d %H:%M')

# combine FL_DATE and CRS_ARR_TIME to get ARR_DATETIME as datetime
flight_data['ARR_DATETIME'] = pd.to_datetime(flight_data['FL_DATE'] + ' ' + flight_data['CRS_ARR_TIME'].astype(str), format='%Y-%m-%d %H:%M')

# combine FL_DATE and DEP_TIME to get ACTUAL_DEP_DATETIME as datetime
flight_data['ACTUAL_DEP_DATETIME'] = pd.to_datetime(flight_data['FL_DATE'] + ' ' + flight_data['DEP_TIME'].astype(str), format='%Y-%m-%d %H:%M')

# combine FL_DATE and ARR_TIME to get ACTUAL_ARR_DATETIME as datetime
flight_data['ACTUAL_ARR_DATETIME'] = pd.to_datetime(flight_data['FL_DATE'] + ' ' + flight_data['ARR_TIME'].astype(str), format='%Y-%m-%d %H:%M')
```

```python
flight_data.info()
```

```python
flight_data.columns
```

```python
flight_data_dep = flight_data[['ORIGIN', 'CRS_DEP_TIME', 'DEP_TIME', 'DEP_DELAY', 'DEP_DATETIME', 'WHEELS_OFF', 'ACTUAL_DEP_DATETIME', 'CRS_ELAPSED_TIME', 'ACTUAL_ELAPSED_TIME', 'AIR_TIME', 'DISTANCE']]
flight_data_arr = flight_data[['ARR_DATETIME', 'ARR_DELAY', 'DEST', 'CRS_ARR_TIME', 'ARR_TIME', 'WHEELS_ON', 'ACTUAL_ARR_DATETIME', 'CRS_ELAPSED_TIME', 'ACTUAL_ELAPSED_TIME', 'AIR_TIME', 'DISTANCE']]
```

```python
airport_data
```

```python
# drop url, county, icao, city_code, time_zone_id, name, elevation
airports_data = airport_data.drop(["url", "county", "icao", "city_code", "time_zone_id", "name", "elevation"], axis=1)
flight_data = flight_data.drop(['FL_DATE', 'TAXI_OUT', 'TAXI_IN'], axis=1)
```

```python
# merge airport_data to flight_data on ORIGIN_AIRPORT_ID
data_join_dep = flight_data_dep.merge(airports_data, left_on='ORIGIN', right_on='code', how='inner')
```

```python
for i in airports_data.columns:
    data_join_dep.rename(columns={i: i+"_origin"}, inplace=True)
```

```python
# merge airport_data to flight_data on ORIGIN_AIRPORT_ID
data_join_arr = flight_data_arr.merge(airports_data, left_on='DEST', right_on='code', how='inner')
```

```python
for i in airports_data.columns:
    data_join_arr.rename(columns={i: i+"_dest"}, inplace=True)
```

```python
# export to pkl
data_join_dep.to_pickle('flight_data_dep.pkl')
data_join_arr.to_pickle('flight_data_arr.pkl')
```

```python
data_gapminder.drop(columns=['drop'], inplace=True)
```

```python
data_gapminder.year = data_gapminder.year.astype(str)
```

```python
# check if the value in lifeExp could be converted to float, if not print it out
for i in data_gapminder.lifeExp:
    try:
        float(i)
    except:
        print(i)

#Â drop all rows in lifeExp == Africa or Asia
data_gapminder = data_gapminder[data_gapminder.lifeExp != 'Africa']
data_gapminder = data_gapminder[data_gapminder.lifeExp != 'Asia']

data_gapminder.lifeExp = data_gapminder.lifeExp.astype(float)
```

```python
data_gapminder
```

```python
# drop NaN
data_gapminder.dropna(inplace=True)
```

```python
data_gapminder
```

```python
data_gapminder.info()
```

```python
# round lifeExp and pop to 2 decimal places
data_gapminder.lifeExp = data_gapminder.lifeExp.round(2)
data_gapminder.pop = data_gapminder["pop"].round(2)

# turn year and gdpPercap into int
data_gapminder.year = data_gapminder.year.astype(int)
data_gapminder.gdpPercap = data_gapminder.gdpPercap.astype(int)
```
